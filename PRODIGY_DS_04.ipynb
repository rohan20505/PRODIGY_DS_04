```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "twitter_validation.csv"
df = pd.read_csv(file_path, header=None)

df.columns = ["ID", "Brand", "Sentiment", "Text"]

print(df.head())
```

```python
# 1. Overall Sentiment Distribution

sentiment_counts = df["Sentiment"].value_counts()

plt.figure()
sentiment_counts.plot(kind="bar")
plt.title("Overall Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
# 2. Sentiment by Brand

brand_sentiment = pd.crosstab(df["Brand"], df["Sentiment"])

brand_sentiment.plot(kind="bar", figsize=(10,6))
plt.title("Sentiment by Brand")
plt.xlabel("Brand")
plt.ylabel("Number of Tweets")
plt.xticks(rotation=45)
plt.legend(title="Sentiment")
plt.tight_layout()
plt.show()
```

```python
# 3. Top Brands with Most Negative Sentiment

negative_by_brand = df[df["Sentiment"] == "Negative"]["Brand"].value_counts().head(10)

plt.figure()
negative_by_brand.plot(kind="bar")
plt.title("Top 10 Brands with Most Negative Sentiment")
plt.xlabel("Brand")
plt.ylabel("Negative Tweet Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Optional: Text length vs sentiment

df["Text_Length"] = df["Text"].astype(str).apply(len)

df.boxplot(column="Text_Length", by="Sentiment", grid=False)
plt.title("Tweet Length by Sentiment")
plt.suptitle("")
plt.xlabel("Sentiment")
plt.ylabel("Text Length")
plt.show()
```
